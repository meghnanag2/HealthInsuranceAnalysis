```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the visual style
sns.set(style="whitegrid")

# 1. Age Distribution by Gender
plt.figure(figsize=(10, 6))
sns.histplot(data=dataset, x="Age", hue="Gender", kde=True, bins=20, palette="Set2", element="step")
plt.title("Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# 2. General Health Status by Gender
plt.figure(figsize=(10, 6))
sns.countplot(data=dataset, x="General_Health", hue="Gender", palette="Set1", order=dataset["General_Health"].value_counts().index)
plt.title("General Health Status by Gender")
plt.xlabel("General Health")
plt.ylabel("Count")
plt.show()

# 3. Total Expenditures by Region
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset, x="Region", y="Total_Expenditures", palette="coolwarm")
plt.title("Total Expenditures by Region")
plt.xlabel("Region")
plt.ylabel("Total Expenditures")
plt.ylim(0, dataset["Total_Expenditures"].quantile(0.95))  # To reduce the influence of extreme outliers
plt.show()

# 4. Family Income vs. Total Self Payment by Employment Status
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x="Family_Income", y="Total_Self_Payment", hue="Employment_Status", alpha=0.6, palette="Dark2")
plt.title("Family Income vs. Total Self Payment by Employment Status")
plt.xlabel("Family Income")
plt.ylabel("Total Self Payment")
plt.show()
```

```python
# COVID-related EDA visualizations

# 1. Expenditure Trends During COVID
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset, x="Year", y="Total_Expenditures", palette="coolwarm")
plt.title("Expenditure Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Total Expenditures")
plt.ylim(0, dataset["Total_Expenditures"].quantile(0.95))  # Clipping extreme outliers for better visibility
plt.xticks(rotation=45)
plt.show()

# 2. Impact of COVID on General Health
health_years = dataset.groupby(["Year", "General_Health"]).size().reset_index(name="Counts")
plt.figure(figsize=(12, 6))
sns.barplot(data=health_years, x="Year", y="Counts", hue="General_Health", palette="Set2")
plt.title("General Health Trends During COVID")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title="General Health", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

# 3. Insurance Plan Trends During COVID
plans_per_year = dataset.groupby("Year")["Person_ID"].nunique().reset_index(name="Unique_Plans")
plt.figure(figsize=(10, 6))
sns.lineplot(data=plans_per_year, x="Year", y="Unique_Plans", marker="o", color="green")
plt.title("Number of Active Insurance Plans Over Time")
plt.xlabel("Year")
plt.ylabel("Unique Insurance Plans")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 4. Smoking Behavior During COVID
smoking_years = dataset.groupby(["Year", "How_Often_Smoke_Cigarettes"]).size().reset_index(name="Counts")
plt.figure(figsize=(12, 6))
sns.barplot(data=smoking_years, x="Year", y="Counts", hue="How_Often_Smoke_Cigarettes", palette="Pastel1")
plt.title("Smoking Behavior Trends During COVID")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title="Smoking Frequency", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

# 5. Expenditures vs. COVID Timeline
plt.figure(figsize=(12, 6))
sns.scatterplot(data=dataset, x="Year", y="Total_Expenditures", hue="How_Often_Smoke_Cigarettes", palette="Dark2", alpha=0.6)
plt.title("Expenditures by Smoking Habits Over Years")
plt.xlabel("Year")
plt.ylabel("Total Expenditures")
plt.ylim(0, dataset["Total_Expenditures"].quantile(0.95))  # Clipping extreme outliers for better visibility
plt.legend(title="Smoking Frequency", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

label_mapping = {
    1: 'Yes',  # Assuming 1 means 'Yes'
    2: 'No'    # Assuming 2 means 'No'
}
# Bar plot of value counts
filtered_dataset = dataset[dataset['Employer_Offers_Health_Insurance'] >= 0]
filtered_dataset['Employer_Offers_Health_Insurance'] = filtered_dataset['Employer_Offers_Health_Insurance'].map(label_mapping)
# Pie chart for proportions
filtered_dataset['Employer_Offers_Health_Insurance'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel')
)
plt.title('Proportion of Employer Offers Health Insurance')
plt.ylabel('')  # Optional to remove the y-label
plt.show()

# Grouped bar chart
cross_tab = pd.crosstab(
    filtered_dataset['Employer_Offers_Health_Insurance'],
    filtered_dataset['Employment_Status']
)
cross_tab.plot(kind='bar', stacked=True, colormap='viridis')

plt.title('Employer Offers Health Insurance by Employment Status')
plt.xlabel('Health Insurance Status')
plt.ylabel('Count')
plt.legend(title='Employment Status')
plt.show()

sns.histplot(
    data=filtered_dataset,
    x='Family_Income',
    hue='Employer_Offers_Health_Insurance',
    multiple='stack',
    palette='viridis'
)
plt.title('Family Income Distribution by Health Insurance Status')
plt.xlabel('Family Income')
plt.ylabel('Count')
plt.show()
```
